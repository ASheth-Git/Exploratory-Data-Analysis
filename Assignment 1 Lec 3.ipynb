Importing the CDK2(CHEMBL301) data.

```python
import pandas as pd
```

```python
data=pd.read_csv("CDK2.csv",sep=";")
```

```python
data
```

**Separating the data based on the "Standard Type"**

```python
IC50=data[data["Standard Type"].str.contains("IC50")]
```

```python
Ki=data[data["Standard Type"].str.contains("Ki")]
```

```python
Kd=data[data["Standard Type"].str.contains("Kd")]
```

```python
data_combined=pd.concat([IC50,Ki,Kd],axis=0)
```

**data_combined is the combined data of IC50 , Ki , Kd**

```python
data_combined
```

```python
data_combined["Molecule ChEMBL ID"].value_counts()
```

**Sorting the Duplicate and Unique data**

```python
Duplicate=data_combined.groupby('Molecule ChEMBL ID').filter(lambda x: len(x)>1)
```

```python
Duplicate['Molecule ChEMBL ID'].value_counts()
```

```python
Unique=data_combined.groupby('Molecule ChEMBL ID').filter(lambda x: len(x)<2)
```

```python
Unique['Molecule ChEMBL ID'].value_counts()
```

```python
data_combined["Molecular Weight"].value_counts()
```

```python
data_new["Mean_Std_Value"]=data_new[['Molecule ChEMBL ID','Smiles','Standard Value']].groupby(["Molecule ChEMBL ID"])["Standard Value"].transform("mean")
```

```python
data_new
```

```python
data_new['Molecule ChEMBL ID'].value_counts()
```

```python
data_new=data_new.sort_values("Mean_Std_Value",ascending=True)
```

```python
data_new
```

```python
data_new=data_new.dropna(axis=0)
```

```python
data_new
```

```python
data_new=data_new.drop_duplicates('Molecule ChEMBL ID',keep='first')
```

```python
data_new
```

```python
data_new['Molecule ChEMBL ID'].value_counts()
```

```python
data_new['Molecular Weight'].value_counts()
```
